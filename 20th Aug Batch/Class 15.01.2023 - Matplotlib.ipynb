```python
x = [1,2,3,4,5]
y = [9,10,4,3,6]
a = [6,8,5,0,3]
b = [7,1,2,9,8]
plt.plot(x,y)
plt.plot(a,b)
plt.title('multi line graph')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x = [1,2,3,4,5]
y = [9,10,4,3,6]
a = [6,8,5,0,3]
b = [7,1,2,9,8]
plt.plot(x,y)
plt.plot(a,b)
plt.title('multi line graph')
plt.show()
```

```python
# Legend
style.use('ggplot')
x = [1,2,3,4,5]
y = [9,10,4,3,6]
a = [6,8,5,0,3]
b = [7,1,2,9,8]
plt.plot(x,y , label = 'plot xy' , color = 'red' , linewidth = 10)
plt.plot(a,b , label = 'plot ab' , color = 'yellow')
plt.legend(loc = 'best')
plt.title('multi line graph')
plt.show()
```
