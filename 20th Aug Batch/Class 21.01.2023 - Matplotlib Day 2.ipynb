```python
#Matplotlib

a = list(range(1,11))
b = list(range(2,12))
c = list(range(3,13))
d = list(range(4,14))

print(a)
print(b)
print(c)
print(d)

```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a,a,'g^',a,b,'bs',a,c,'r--',a,d,'mo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Multiple lines with shapes and colours')
plt.show()
```

```python
list1 = [10,30,20,22,35,70,56]
plt.pie(list1,labels = list1)
plt.show()
```

```python
#Violin Plot

plt.violinplot([10,11,16,34,18,12,18,18,18,19,21,20])
plt.show()
```
