#### One Hot Encoder

```python
import numpy
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'team':['A','A','B','B','B','B','C','C'],
                   'points': [25,12,26,74,19,56,29,35]})
```

```python
df.head()
```

```python
oh = OneHotEncoder()
```

```python
oh_df = pd.DataFrame(oh.fit_transform(df[['team']]).toarray())
```

```python
oh_df.head()
```

```python
df1 = df.join(oh_df)
```

```python
df1.head()
```

```python
df1.drop('team',axis=1,inplace = True)
```

```python
df1.head()
```

```python
df1.columns = ['points','TeamA','TeamB','TeamC']
```

```python
df1.head()
```

#### Linear Regression Implementation

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets , linear_model
```

```python
diabetes = datasets.load_diabetes()
```

```python
print(diabetes.keys())
```

```python
print(diabetes.data)
```

```python
print(diabetes.DESCR)
```

```python
print(diabetes.feature_names)
```

```python
print(diabetes.target)
```

```python
diabetes_x = diabetes.data[:,np.newaxis,2]
```

```python
print(diabetes_x)
```

```python
print(len(diabetes_x))
```

```python
print(diabetes_x.shape)
```

```python
diabetes_x_train = diabetes_x[:-20]
diabetes_x_test = diabetes_x[-20:]
diabetes_x_train.shape
diabetes_x_test.shape
```

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
diabetes_y_train.shape
diabetes_y_test.shape
```

```python
lr_model = linear_model.LinearRegression()
```

```python
lr_model.fit(diabetes_x_train,diabetes_y_train)
```

```python
lr_model.score(diabetes_x_test,diabetes_y_test)
```
