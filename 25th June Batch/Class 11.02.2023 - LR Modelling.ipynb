#### One Hot Encoder

```python
import numpy
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'team':['A','A','B','B','B','B','C','C'],
                   'points': [25,12,26,74,19,56,29,35]})
```

```python
df.head()
```

```python
oh = OneHotEncoder()
```

```python
oh_df = pd.DataFrame(oh.fit_transform(df[['team']]).toarray())
```

```python
oh_df.head()
```

```python
df1 = df.join(oh_df)
```

```python
df1.head()
```

```python
df1.drop('team',axis=1,inplace = True)
```

```python
df1.head()
```

```python
df1.columns = ['points','TeamA','TeamB','TeamC']
```
